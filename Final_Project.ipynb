## Research Question
**Research Question:** Does social media engagement and product price affect purchase behavior (e.g., number of purchases)?

## Data Overview
In this section, we prepare our dataset for analysis by simulating an e-commerce environment. We will display the first few rows, structural information, and summary statistics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for academic and clean Look
sns.set_theme(style="whitegrid", palette="muted")

# 1. Create a synthetic dataset
np.random.seed(42)
n_samples = 500

data = {
    'product_id': range(1, n_samples + 1),
    'category': np.random.choice(['Electronics', 'Clothing', 'Home', 'Beauty', 'Sports'], n_samples),
    'price': np.random.uniform(10, 500, n_samples),
    'engagement_score': np.random.normal(50, 15, n_samples), # Social media engagement (e.g., likes, comments)
    'sentiment_index': np.random.uniform(0, 1, n_samples),   # 0 = Negative, 1 = Positive
}

df = pd.DataFrame(data)

# Introduce a relationship where lower price and higher engagement = more purchases
base_purchase = np.random.poisson(20, n_samples)
df['purchases'] = (base_purchase 
                   + (df['engagement_score'] * 0.5) 
                   - (df['price'] * 0.05) 
                   + (df['sentiment_index'] * 10)).astype(int)
df['purchases'] = np.clip(df['purchases'], 0, None)  # Ensure no negative purchases

print("Data Head:")
display(df.head())

print("\nData Info:")
df.info()

print("\nData Description:")
display(df.describe())
```

## Univariate Analysis
In this section, we explore the individual distributions of key variables in our dataset to understand their overall behavior before analyzing relationships. We have chosen to look at the distribution of product prices and the categorical distribution of product categories.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Product Prices', fontsize=16)
plt.xlabel('Price ($)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.tight_layout()
plt.show()
```

**Explanation:** The histogram above illustrates the distribution of product prices in our dataset. The prices are roughly uniformly distributed between $10 and $500, representing a wide variety of product offerings without a dominant price cluster.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='category', order=df['category'].value_counts().index, palette='viridis')
plt.title('Distribution of Products by Category', fontsize=16)
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.tight_layout()
plt.show()
```

**Explanation:** This bar chart shows the count of products available in each category. We can observe the dataset is relatively balanced across Electronics, Beauty, Home, Clothing, and Sports, ensuring our subsequent analysis isn't heavily skewed toward a single category.

## Bivariate Analysis
Next, we analyze the relationships between our target variable (`purchases`) and potential predictors (`engagement_score` and `price`). This will directly help us answer our core research question.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='engagement_score', y='purchases', alpha=0.6, color='coral')
plt.title('Product Purchases vs. Social Media Engagement', fontsize=16)
plt.xlabel('Engagement Score', fontsize=14)
plt.ylabel('Number of Purchases', fontsize=14)
plt.tight_layout()
plt.show()
```

**Explanation:** The scatter plot demonstrates a positive correlation between the social media engagement score and the number of purchases. Higher engagement generally aligns with increased purchase volume, validating that social activity translates to sales.

```python
plt.figure(figsize=(10, 6))
# Binning prices to create readable boxplots
df['price_tier'] = pd.qcut(df['price'], q=4, labels=['Low', 'Medium-Low', 'Medium-High', 'High'])

sns.boxplot(data=df, x='price_tier', y='purchases', palette='Set2')
plt.title('Purchases Distribution Across Price Tiers', fontsize=16)
plt.xlabel('Price Tier', fontsize=14)
plt.ylabel('Number of Purchases', fontsize=14)
plt.tight_layout()
plt.show()
```

**Explanation:** This box plot groups products into four price tiers and shows the distribution of purchases for each. We can observe a distinct negative trend where products in higher price tiers generally have significantly fewer purchases compared to lower-priced items.

## Insights
Based on the visualizations above, we can draw the following conclusions to answer our research question:

- **Engagement impacts purchase:** There is a clear positive trend showing that products with higher social media engagement scores receive a larger volume of purchases (as seen in the scatter plot).
- **Price is a limiting factor:** When analyzing price tiers, lower-priced items see significantly higher median purchases compared to more expensive items, showing an inverse relationship (as seen in the box plot).

**Final Answer:** Yes, social media engagement positively affects purchase activity, but it must be contextualized with the product's price. More expensive products tend to see lower purchase volumes despite similar engagement levels.
